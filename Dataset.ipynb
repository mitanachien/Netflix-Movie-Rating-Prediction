```python
count = 0
training = []
testing = []
for id in unique_ID:
  condition = df['Customer Id']==id
  filtered_df = df[condition]
  new_df = filtered_df.reset_index()
  id_len = len(new_df)
  data_len = int(id_len * 0.5)
  training_id = new_df.loc[:data_len]
  testing_id = new_df.loc[data_len+1:]
  training.append(training_id)
  testing.append(testing_id)
  count+=1
  if count%5000 == 0:
    print('Process {0:d} times'.format(count))
training_data = pd.concat(training, ignore_index=True)
testing_data = pd.concat(testing, ignore_index=True)
```
